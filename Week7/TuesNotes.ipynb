```python
import math
import numpy as np
from scipy.integrate import odeint
odeint?
```

```python
def derive(Theta,t):
    g = 9.8
    l = .5
    theta = Theta[0]
    omega = Theta[1]
    dtheta = omega
    domega = (-g/l)*sin(theta)
    return array([dtheta,domega])
```

```python
g = 9.8
l = .5
th0 = 20*pi/180
t = arange(0.,10+.01,.01)
theta_apt = lambda t: th0*cos(sqrt(g/l)*t)
omega0 = 0.
theta_init = array([th0, omega0])

theta = odeint(derive,theta_init,t)
```

```python
plot(t,theta_apt(t),'b-',t,theta[:,0],'r--')
show()
```
