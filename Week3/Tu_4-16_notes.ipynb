Create a very simple plot

```python
t = arange(0.0,1.0+0.01,0.01)
c = cos(2*2*pi*t)
s = sin(4*pi*t)
```

```python
plot(t,c, color = "violet",linewidth = 3,linestyle=":", label="$cos(4{\pi}t)$")
plot(t,s, color = "green",linewidth = 1.5,linestyle="--", label="$sin(4{\pi}t)$")
xlabel('time (s)')
ylabel('volatage (mV)')
#Add title for convenience
title('A simple plot',fontsize=48)
grid(False)
legend(loc = "best")
savefig('plot.pdf')
```
