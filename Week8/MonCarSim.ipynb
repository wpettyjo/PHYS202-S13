```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.

plot(x,f(x),label='True Distribution',lw=3)
plot(x,F(x),label='Cumulitive Function',lw=3)

xlim = (0.,3.5)
ylim = (0.,1.1)

legend(loc=0)
show()
```

```python
#number samples
N = 10000

#N random numbers betweem 1 and 0
z = numpy.random.random_sample(N)

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#sampled x values
xsamples = Finv(z)

#how do they compare to the original?
hist(xsamples,50,normed=True,label='Sampled Distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

$f(x) = \lambda e^{-\lambda x}$

```python
L = 1
x = arange(0.,5.,.01)
f = lambda x:L * e**(-L*x)
F = lambda x:1 + (-L * e**(-L*x))
plot(x,f(x),label='Real',lw=3,color='red')
plot(x,F(x),label='CDF',lw=3)

show()
```

```python
N = 10000
z = numpy.random.random_sample(N)
Finv = lambda z: -log(1-z)
xsamples = Finv(z)
hist(xsamples,50,normed=True,label='Sampled Distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')
```
