```python
import numpy as np

def pointpotential(x,y,q,Xo,Yo):
    k = 8.9876628 * 10**9
    return (k*q)/np.sqrt((x-Xo)**2 - (y-Yo)**2)

def dipoleppotential(x,y,q,d):
    return pointpotential(x,y,q,-d,0) + pointpotential(x,y,-q,d,0)

def pointfield(x,y,q,Xo,Yo):
    k = 8.9876628 * 10**9
    Ex = ((x - Xo)*(k*q))/np.sqrt((x-Xo)**2 - (y-Yo)**2)
    Ey = ((y - Yo)*(k*q))/np.sqrt((x-Xo)**2 - (y-Yo)**2)
    return Ex
    return Ey
```

```python
def is_anagram(s1,s2):
    st1 = []
    st2 = []
    if len(s1) == len(s2):
        for i in s1:
            i = i.lower()
            st1.append(i)
        for n in s2:
            n = n.lower()
            st2.append(n)
        for q in st1:
            if q == st2[:]:
                print True 
            else:
                return False
            
```
