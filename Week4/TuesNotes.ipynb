Electric Dipole Potential

$V(x,y) = \frac{kq}{\sqrt{x^2 + (y - d)^2}} - \frac{kq}{\sqrt{x^2 + (y +d)^2}}$

```python
k = 1.
q = 1.
d = 1.
V = ((k*q)/((x**2. + (y-d)**2.)**(1./2.))) - ((k*q)/((x**2. + (y+d)**2.)**(1./2.))) 
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.copper)

cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(x, y, V, zdir='z', offest=-40., cmap=cm.copper)
```

determine electric potental for a quadropole consisting of q = -1.0 nC at (x+-2,y=0) & q = 1.0nC at (x=0,y+-2)
range = -4,4 101 bins
$k = 8.9875518 * 10^9 \frac{Nm^2}{C^2}$

```python
dx = .1
dy = .1
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
x,y = meshgrid(x,y)
```

```python
k = 8.9875518 * 10.**9.
q1 = 1.
q2 = -1.
d = 1.
V = ((k*q1*q2)/((x**2. + (y-d)**2.)**(1./2.))) - ((k*q1*q2)/((x**2. + (y+d)**2.)**(1./2.))) 
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
pl = ax.plot_surface(x,y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.ocean)
```
