```python
f = lambda x: x**(.5) + sin(x)
```

```python
x = arange(0.,15.,0.001)
```

```python
theplot = plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
xlabel(r'$x$',fontsize = 20)
xlim(0.,15.)

show()
```

```python
xmin = 1.7
xmax = 12.4

subx = x[logical_and((x>xmin),(x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0

theplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
```

```python
N = 100000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);

theplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:10000],samples_y[1:10000],'.')

plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')

show()
```

```python
newmask = (samples_y < f(samples_x))
```

```python
theplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:100000],samples_y[1:100000],'.')

plot(samples_x[newmask[:1000000]],samples_y[newmask[:1000000]],'r.')

plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')

show()
```

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

```python
Int = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Int(xmax) - Int(xmin)
Ierr = abs(100 * (Iexact - I)/Iexact)
print Iexact
print I
print Ierr
```

```python
g = lambda x: x**(.5) + sin(x)
```

```python
xmax = 12.4
xmin = 1.7
```

```python
n = 10000
x_samples = xmin + (xmax-xmin)*random_sample(n)
sigG = g(x_samples)
print sigG
```

```python
IntAve = sum(sigG)*(xmax-xmin)/n
print IntAve
```

```python
plotme = sum(sigG)/n
```

```python
myplot = plot(x,g(x),lw=3)

plot([xmin,xmax],[plotme, plotme], color = 'r',lw = 3)
```

```python
Int = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Int(xmax) - Int(xmin)
Ierr = abs(100 * (Iexact - IntAve)/Iexact)
print Iexact
print IntAve
print Ierr
```
